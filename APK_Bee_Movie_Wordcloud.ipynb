```python
#For this assignment I decided to create a word cloud for the Bee Movie Script 
import requests
response = requests.get('https://web.njit.edu/~cm395/theBeeMovieScript/') 
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content, 'html5lib')
text = soup.get_text(strip=True)
```

```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stops = stopwords.words('english')
```

```python
from textblob import TextBlob
```

```python
blob = TextBlob(text)
[word for word in blob.words if word not in stops]
```

```python
import imageio
mask_image = imageio.imread('mask_circle.png')
```

```python
from wordcloud import WordCloud
```

```python
wordcloud = WordCloud(width=1000, height=1000, 
    colormap='prism', mask=mask_image, background_color='white')
```

```python
wordcloud = wordcloud.generate(text)
```

```python
wordcloud = wordcloud.to_file('BeeMovieWordCloud.png')
```
